```python
!pip install qiskit qiskit-aer --quiet
from qiskit import QuantumCircuit
qc = QuantumCircuit (2,2) # 2 qbits, 2 Classical bits
qc.h(0) #Hadamard gate. Sets qbits into superposition
qc.cx (0,1) # CNOT gate (control = 0, target=1)
qc.measure([0,1],[0,1]) # Measure qbits into classial bits
qc.draw() #Visulaize circuit
```
